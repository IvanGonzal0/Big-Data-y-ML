```python
import plotly.graph_objects as go
import pandas as pd
```

```python
path = '/content/Dataset_ventas_AbrilT.csv'
```

```python
df_ventas = pd.read_csv(path)
df_ventas
```

```python
# Transformo todos los valores no numericos en Nan en las columnas Cantidad Pedida y Precio Unitario
df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')

# Elimino todos los Nan y me quedo solo con los valores numericos
df_ventas = df_ventas.dropna()

# Transformo todos los valores numericos al tipo de dato int
df_ventas['Cantidad Pedida'] = df_ventas['Cantidad Pedida'].astype(int)
df_ventas['Precio Unitario'] = df_ventas['Precio Unitario'].astype(float)
df_ventas
```

```python
Productos_vendidos = df_ventas.groupby('Producto')['Cantidad Pedida'].sum() #sumo los productos y la cantidad pedida, para obtener el total de producto vendido
Productos_vendidos_ordenados = Productos_vendidos.sort_values(ascending = False)  #los ordeno para que queden arriba los mas vendidos
Productos_vendidos_ordenados
```

```python
#voy a crear un nuevo dataset limpio para graficar
df_productos = pd.DataFrame({'Producto':Productos_vendidos_ordenados.index, 'Cantidad Pedida': Productos_vendidos_ordenados.values})

df_productos
```

```python
#usando plotly

productos = df_productos['Producto']
unidades = df_productos['Cantidad Pedida']
#creo un objeto figura usando plotly graphic_object (go)
fig = go.Figure()

#Agrego el grafico en el objeto figura creado
fig.add_trace(
    go.Bar(
        x= productos,
        y= unidades,
        name = 'Unidades Vendidas',
        marker = dict(color = unidades, colorscale = 'RdYlGn', colorbar = dict(title='Total de ventas'))  #el marker se vincula a todo lo graficado, en cuanto a color estilo variables tomadas en cuenta
    )
);

#puedo modificar la figura a gusto
fig.update_layout(
    title='Unidades vendidas',
    xaxis= dict(title='Productos'),
    yaxis= dict(title='Unidades')
)
```

```python
#la funcion .agg nos permite hacer multiples operaciones matematicas o de lo que sea
#Extraigo valores del df
analisis_de_productos = df_ventas.groupby('Producto').agg({
    'Cantidad Pedida':'sum',
    'Precio Unitario': 'mean'
})

analisis_de_productos
#ME QUEDA PRODUCTO DE INDICE
```

```python
#necesito que se ordenen y se reemplace el indice por un indice de verdad

analisis_de_productos = analisis_de_productos.sort_values(by='Cantidad Pedida', ascending=False)

#Para resetear el indice utilizo reset_index(inplace=True)
analisis_de_productos.reset_index(inplace=True)
analisis_de_productos
```

```python
analisis_de_productos.index
```

```python
#usando plotly

productos = analisis_de_productos['Producto']
unidades = analisis_de_productos['Cantidad Pedida']
precio = analisis_de_productos['Precio Unitario']
#creo un objeto figura usando plotly graphic_object (go)
fig = go.Figure()

#Agrego el grafico en el objeto figura creado
fig.add_trace(
    go.Bar(
        x= productos,
        y= unidades,
        name = 'Unidades Vendidas',
        marker = dict(color = 'Light Blue'),  #el marker se vincula a todo lo graficado, en cuanto a color estilo variables tomadas en cuenta
        yaxis = 'y'
    )
);

#Puedo agregar una nueva figura para mostrar la variacion de precios segun el producto utilizando scater

fig.add_trace(go.Scatter(
            x= productos,
        y= precio,
        name = 'Unidades Vendidas',
        line = dict(color = 'Red', shape='spline', smoothing=1.3, width=2),
        yaxis = 'y2'
))


#puedo modificar la figura a gusto
fig.update_layout(
    title='Unidades vendidas y precio de productos',
    xaxis= dict(title='Productos'),
    yaxis= dict(title='Unidades', side='left'),
    yaxis2 = dict(title = 'Precio', overlaying='y',side = 'right'),
    yaxis_range =[0, max(unidades)*1.1],
    yaxis2_range = [0, max(precio)*1.1]
)
```

```python
#vamos a separar la fecha y hora, y vamos a obtener la fecha y hora en otra columna de type datetime
#transformo los valores fecha hora a un tipo datetime
df_ventas['Fecha de compra'] = pd.to_datetime(df_ventas['Fecha de Pedido'])

df_ventas
```

```python
# Ahora vamos a extrar datos del dato 'Fecha de compra'

def extraer_datos(datos):
  #aca agarro el dataset y creo una nueva columna con la hora, usando el dato datetime que cree antes en la columna Fecha de compra
  datos['Hora'] = datos['Fecha de compra'].dt.hour

  return datos

df_ventas = extraer_datos(df_ventas)
df_ventas
```

```python
#creo las variables de x e y respectivamente(horas y ventas_por_hora)
horas = list(range(24))
ventas_por_hora = df_ventas.groupby('Hora')['Cantidad Pedida'].sum()

#creamos la figura
fig = go.Figure()

#creamos el grafico scatter
fig.add_trace(go.Scatter(
    x=horas,
    y=ventas_por_hora,
    name = 'Ventas por hora'
))

#modificamos los titles y los axis
fig.update_layout(
    title= 'Unidades vendidas por hora',
    xaxis= dict(title='Hora', tickvals = horas),
    yaxis= dict(title='Unidades Vendidas')
)
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=horas,
    y=ventas_por_hora,
    name = 'Ventas por hora'
))

fig.update_layout(
    title= 'Unidades vendidas por hora',
    xaxis= dict(title='Hora', tickvals = horas),
    yaxis= dict(title='Unidades Vendidas')
)
```

```python
#Ahora vamos a ver que dias de la semana se vende mas
# Ahora vamos a extrar datos del dato 'Fecha de compra'

def extraer_datos(datos):
  #aca agarro el dataset y creo una nueva columna con la hora, usando el dato datetime que cree antes en la columna Fecha de compra
  datos['Hora'] = datos['Fecha de compra'].dt.hour
  datos['Dia'] = datos['Fecha de compra'].dt.day
  return datos

df_ventas = extraer_datos(df_ventas)
df_ventas
```

```python
#creo las variables de x e y respectivamente(horas y ventas_por_hora)
#dias = list(range(32))

ventas_por_dia = df_ventas.groupby('Dia')['Cantidad Pedida'].sum()
dias = list(ventas_por_dia.index)
#creamos la figura
fig = go.Figure()

#creamos el grafico scatter
fig.add_trace(go.Scatter(
    x=dias,
    y=ventas_por_dia,
    name = 'Ventas por hora',
    line = dict(color = "Green", shape = 'spline', smoothing=1.3,width=3)
))

#modificamos los titles y los axis
fig.update_layout(
    title= 'Unidades vendidas por dia',
    xaxis= dict(title='Hora', tickvals = dias),
    yaxis= dict(title='Unidades Vendidas'),
    #yaxis_range = [0, max(ventas_por_dia)*1.1]
)
```

```python
# Analizamos las ventas por estado, como loseparamos?

direccion_df = df_ventas['Dirección de Envio'].str.split(', ',expand=True)



#estado_df=direccion_df['2']
#separamos el codigo postal de estado, en codigo postal y estado
direccion_df.columns=['Calle', 'Ciudad' , 'Cod Postal Estado']
direccion_df
direccion_df[['Estado', 'Cod Postal']] = direccion_df['Cod Postal Estado'].str.split(' ', expand=True)
direccion_df
```

```python
#ahora vamo,s a hacer un df desglosado

df_ventas_desglosado = pd.concat([df_ventas, direccion_df], axis=True)  #esta funcion me une el df original con el que tiene las direcciones que creamos anteriormente, y el True me respeta los indices

df_ventas_desglosado
```

```python
#ahora agrupamos ventas por estado

ventas_por_estado = df_ventas_desglosado.groupby('Estado')['Cantidad Pedida'].sum()
#aca ago el indice para este df nuevo agrupado por Estado y sumando la cantidad de pedidos (por estado tambien)
ventas_por_estado = ventas_por_estado.reset_index(inplace=False)

ventas_por_estado
```

```python
import plotly.express as px
#vamos a trabajar con EEUU, pero se puede usar el mapa mundi, o la region que querramos
fig = px.choropleth(ventas_por_estado, locations='Estado', locationmode='USA-states', color='Cantidad Pedida',
                    color_continuous_scale='RdYlGn',
                    scope='usa',
                    title='Ventas por estado'
)

fig.show()
```
