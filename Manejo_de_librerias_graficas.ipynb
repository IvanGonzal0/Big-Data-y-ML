```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
```

```python
# Cargo un dataset ya existente
ruta_de_acceso = '/content/drive/MyDrive/UPSO ML & BD/Dataset de ventas2/Dataset_de_ventas_Abril.csv'

df_ventas = pd.read_csv(ruta_de_acceso)

df_ventas
```

```python
# Eliminamos las filas que contienen NaN
df_ventas = df_ventas.dropna()

# Cambia los Nan por 0
df_ventas = df_ventas.fillna(0)

df_ventas = df_ventas.fillna({'Cantidad Pedida':0,'Precio Unitario':2})
```

```python
# Transformo todos los valores no numericos en Nan en las columnas Cantidad Pedida y Precio Unitario
df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')

# Elimino todos los Nan y me quedo solo con los valores numericos
df_ventas = df_ventas.dropna()

# Transformo todos los valores numericos al tipo de dato int
df_ventas['Cantidad Pedida'] = df_ventas['Cantidad Pedida'].astype(int)
df_ventas['Precio Unitario'] = df_ventas['Precio Unitario'].astype(float)

```

```python
# Agrupo los Productoos y la cantidad ordenada y realizo la sumatoria de los Productoos vendidos
Productoos_vendidos = df_ventas.groupby('Producto')['Cantidad Pedida'].sum() #Aca se puede hacer muchas operaciones .sum() .min() .max() .std() etc

# Ordeno los de mayor a menor segun la cantidad de Productoos vendidos
Productoos_vendidos_ordenados = Productoos_vendidos.sort_values(ascending=False )

df_Productoos = pd.DataFrame({'Producto':Productoos_vendidos_ordenados.index, 'Cantidad Pedida':Productoos_vendidos_ordenados.values})
```

## Grafico de barras


```python
import plotly.graph_objects as go

# Extraigo valores del dataFrame para realizar los graficos
Productoos = df_Productoos['Producto'].values
Unidades = df_Productoos['Cantidad Pedida'].values

# Creo el objeto figura
fig = go.Figure()


# Agrego al objeto figura un grafico de barra conn los valores extraidos previamente
fig.add_trace(go.Bar(
        x= Productoos,
        y= Unidades,
        yaxis="y",
        name="Unidades vendidas",
        marker=dict(color="light blue")))


# Configuro titulos y ejes del grafico
fig.update_layout(
    title="Unidades vendidas",
    xaxis=dict(title="Productoo"),
    yaxis=dict(title="Unidades vendidas", side="left"))


# Muestro el grafico
fig.show()
```

## Grafico multiple

```python
# Extraigo valores del dataFrame para realizar el grafico
Analisis_de_productos = df_ventas.groupby('Producto').agg({
    'Cantidad Pedida':'sum',
    'Precio Unitario': 'mean'
})

# Ordeno de mayor a menor
Analisis_de_Productoos_ordenados = Analisis_de_productos.sort_values(by='Cantidad Pedida', ascending = False)

# Reseteo Indice
Analisis_de_Productoos_ordenados = Analisis_de_productos.reset_index(inplace=False)

Analisis_de_Productoos_ordenados
```

```python
import plotly.graph_objects as go

# Extraigo valores del dataFrame para realizar los graficos
Productoos = Analisis_de_Productoos_ordenados['Producto'].values
Unidades = Analisis_de_Productoos_ordenados['Cantidad Pedida'].values
Precios =  Analisis_de_Productoos_ordenados['Precio Unitario'].values

# Creo el objeto figura
fig = go.Figure()

# Agrego al objeto figura un grafico de barra con los valores extraidos previamente de Productoos y unidades
fig.add_trace(go.Bar(

        x= Productoos,
        y= Unidades,
        yaxis="y",
        name="Unidades vendidas",
        marker=dict(color="light blue")))


# Agrego al objeto figura un grafico de de linea conn los valores extraidos previamente de Productoos y precios
fig.add_trace(go.Scatter(

        x= Productoos,
        y= Precios,
        mode='lines',
        name="Precio",
        yaxis="y2",
        line=dict(color="red", shape='spline', smoothing=1.3, width=3)))  # Realizo ajustes para la línea suave y ancha

# Configuro los ejes y los títulos
fig.update_layout(

        title="Unidades vendidas y Precio por Productoo",
        xaxis=dict(title="Productoo"),
        yaxis=dict(title="Unidades vendidas", side="left"),
        yaxis2=dict(title="Precio", overlaying="y", side="right"),
        yaxis_range=[0, max(Unidades) * 1.1],  # Ajusto los rangos de los ejes para que ambos comiencen en cero
        yaxis2_range=[0, max(Precios) * 1.1])


fig.show()
```

## **PREGUNTA:** ¿A qué hora debemos mostrar la publicidad para maximizar la probabilidad de que el cliente compre el Productoo?

```python
df_ventas
```

```python
# Trasformo los valores en un formato de fecha y hora
df_ventas['Fecha de Pedido'] = pd.to_datetime(df_ventas['Fecha de Pedido']) #Si no aplicamos esto, cuando queremos extraer la hora nos tira error
```

```python
# Realizamos una funcion
def extraer_horario(datos):
  datos['Hora'] = datos['Fecha de Pedido'].dt.hour

  return datos
```

```python
df_ventas = extraer_horario(df_ventas) #Aplicamos la funcion

df_ventas['Precio Total'] =  df_ventas['Cantidad Pedida'] * df_ventas['Precio Unitario']

df_ventas
```

```python
# Preparo eje x que va de 0 a 23

horas =  list(range(24))
```

```python
import plotly.graph_objects as go

Horas = list(range(24))
Ventas_por_hora = df_ventas.groupby('Hora')['Cantidad Pedida'].count()

# Creo el objeto figura
fig = go.Figure()


# Agrego al objeto figura un grafico de de linea con las horas y las ventas por hora
fig.add_trace(go.Scatter(

    x=Horas,
    y=Ventas_por_hora,
    mode='lines',
    name="Ventas por hora",
    line=dict(color="red", shape='spline', smoothing=1.3, width=3)))


# Configuro los ejes y los títulos
fig.update_layout(

    title="Unidades vendidas por hora",
    xaxis=dict(title="Hora", tickvals=Horas),  # Para que en el eje x se muestren todos los valores uso tickvals=Horas
    yaxis=dict(title="Unidades vendidas", side="left"))

fig.show()

```

**RESPUESTA:** Para maximizar la probabilidad de que el cliente compre el Productoo, se recomienda mostrar la publicidad en dos momentos clave: alrededor del mediodía, aproximadamente a las 12:00 hs, y durante la tarde a las 19:00 hs. Estos momentos han demostrado tener un mayor impacto en las decisiones de compra, lo que sugiere una mayor receptividad por parte de los clientes en esos intervalos horarios.

```python
def extraer_datos(datos):

  datos['Hora'] = datos['Fecha de Pedido'].dt.hour
  datos['Dia'] = datos['Fecha de Pedido'].dt.day
  return datos


df_ventas = extraer_datos(df_ventas)
```

```python

Compras_dia = df_ventas.groupby('Dia')['Cantidad Pedida'].count().astype('int')
Dias = list(Compras_dia.index)

# Creo el objeto figura
fig = go.Figure()


# Agrego al objeto figura un grafico de de linea con las horas y las ventas por hora
fig.add_trace(go.Scatter(

    x=Dias,
    y=Compras_dia,
    mode='lines',
    name="Ventas por hora",
    line=dict(color="red", shape='spline', smoothing=1.3, width=3)))


# Configuro los ejes y los títulos
fig.update_layout(

    title="Unidades vendidas en el mes de abril",
    xaxis=dict(title="Dia", tickvals=Dias),  # Para que en el eje x se muestren todos los valores uso tickvals=Horas
    yaxis=dict(title="Unidades vendidas", side="left"))

fig.show()
```

```python
address_df =df_ventas['Dirección de Envio'].str.split(', ', expand=True)
address_df.columns = ['Calle', 'Ciudad', 'Cod Postal Estado']
address_df[['Estado', 'Cod Postal']] = address_df['Cod Postal Estado'].str.split(' ', expand=True)
df_ventas_desglosado = pd.concat([df_ventas, address_df], axis=1)

# TOTAL SALES BY CITY
ventas_por_ciudad = df_ventas_desglosado.groupby('Ciudad').agg({'Cantidad Pedida': 'sum', 'Precio Total': 'sum'}).reset_index()

```

```python
ventas_por_ciudad=ventas_por_ciudad.sort_values(by='Precio Total',ascending=False)

ventas_por_ciudad
```

```python
import plotly.graph_objects as go

# Extraigo valores del dataFrame para realizar los graficos
Ciudades = ventas_por_ciudad['Ciudad'].values
Unidades = ventas_por_ciudad['Cantidad Pedida'].values
Precios =  ventas_por_ciudad['Precio Total'].values

# Creo el objeto figura
fig = go.Figure()

# Agrego al objeto figura un grafico de barra con los valores extraidos previamente de Productoos y unidades
fig.add_trace(go.Bar(

        x= Ciudades,
        y= Unidades,
        yaxis="y",
        name="Unidades vendidas",
        marker=dict(color="light blue")))


# Agrego al objeto figura un grafico de de linea conn los valores extraidos previamente de Productoos y precios
fig.add_trace(go.Scatter(

        x= Ciudades,
        y= Precios,
        mode='lines',
        name="Precio",
        yaxis="y2",
        line=dict(color="red", shape='spline', smoothing=1.3, width=3)))  # Realizo ajustes para la línea suave y ancha

# Configuro los ejes y los títulos
fig.update_layout(

        title="Unidades vendidas y Precio por Productoo",
        xaxis=dict(title="Productoo"),
        yaxis=dict(title="Unidades vendidas", side="left"),
        yaxis2=dict(title="Precio", overlaying="y", side="right"),
        yaxis_range=[0, max(Unidades) * 1.1],  # Ajusto los rangos de los ejes para que ambos comiencen en cero
        yaxis2_range=[0, max(Precios) * 1.1])


fig.show()
```
