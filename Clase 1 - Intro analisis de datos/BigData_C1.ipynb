```python
# IMPORTANDO CSV

path = '/content/Dataset_de_ventas_Abril.csv'

df_ventas = pd.read_csv(path)

df_ventas
```

```python
df_ventas.head()
```

```python
# Limpiar el dataset de valores NaN

df_ventas = df_ventas.dropna()
df_ventas.head(10)
```

```python
# Limpiar NaN reemplazando todos los Nan por 0

df_ventas = df_ventas.fillna(0)
df_ventas
```

```python
# Otra forma de manipular los NaN
df_ventas = df_ventas.fillna({'Quantity Ordered' : 0, 'Price Each':2})
df_ventas
```

```python
# Otra vez para sacar los NaN definitivamente (es lo mas comun)

df_ventas = df_ventas.dropna()
df_ventas
```

```python
# Como reemplazar DATOS que no sean numericos a NaN (por que esten mal cargados), con el errors= 'coerce' transformo todos esos datos que no sean numeros en NaN
# Aca todos los valores de quantity order y price each pasamelos a NaN para luego limpiar el df de nuevo
# Esto me sirve para poder analizar sin problemas y sin errores (LIMPIEZA)

df_ventas['Quantity Ordered'] = pd.to_numeric(df_ventas['Quantity Ordered'] ,errors = 'coerce')
df_ventas['Price Each'] = pd.to_numeric(df_ventas['Price Each'] ,errors = 'coerce')

df_ventas.head()
```

```python
# limpio de nuevo los NaN por caso que hubiera valores que no eran numeros

df_ventas = df_ventas.dropna()
df_ventas.head()
```

```python
# Para clonar un df podemos hacer

df_ventas2 = df_ventas.copy()
```

```python
# LIMPIEZA y manipulacion de tipos de datos

df_ventas['Quantity Ordered'] = pd.to_numeric(df_ventas['Quantity Ordered'] ,errors = 'coerce')
df_ventas['Price Each'] = pd.to_numeric(df_ventas['Price Each'] ,errors = 'coerce')

df_ventas = df_ventas.dropna()

df_ventas['Quantity Ordered'] = df_ventas['Quantity Ordered'].astype(int)
df_ventas['Price Each'] = df_ventas['Price Each'].astype(float)

df_ventas.head()
```

```python
# Filtramos los productos que se pidieron mas de 5 unidades en la misma orden de compra
df_alto_valor = df_ventas.loc[df_ventas['Quantity Ordered']>=5]
df_alto_valor
```

```python
# Podemos agrupar los productos
productos_vendidos = df_ventas.groupby('Product')['Quantity Ordered']

productos_vendidos.head()
```

```python
# Podemos agrupar los productos y sumar las cantidades
productos_vendidos = df_ventas.groupby('Product')['Quantity Ordered'].sum()

productos_vendidos.head()
```

```python
# Podemos agrupar sumar y ordenarlos

productos_vendidos = df_ventas.groupby('Product')['Quantity Ordered'].sum()
productos_vendidos = productos_vendidos.sort_values(ascending = False)
productos_vendidos.head()
```

```python
# Analisis de productos
# agg es ucnaod quiero hacer multiples operaciones ( se usa un diccionario y se le pasa una clave valor con columna : funcion a realizar)
analisis_de_productos = df_ventas.groupby('Product').agg({
    'Quantity Ordered':'sum',
    'Price Each': 'mean'
})

analisis_de_productos = analisis_de_productos.sort_values(by ='Quantity Ordered', ascending = False)

analisis_de_productos
```

```python
# Aca puedo obtener el total de ventas de todos los productos haciendo la cuenta de la cantidad de productos vendidos por orden * el precio de cada producto
analisis_de_productos['Total Sales'] = analisis_de_productos['Quantity Ordered'] *analisis_de_productos['Price Each']

# obtenemos el promedio de prodcutos*precio
analisis_de_productos
```

```python
#aca lo podemos ordenar por la ganancia genereda. a partir la columna total sales
analisis_de_productos = analisis_de_productos.sort_values(by='Total Sales', ascending = False)
analisis_de_productos
```

```python
# Aca ordeno por la mayor cantidad de productos vendidos
analisis_de_productos = analisis_de_productos.sort_values(by='Quantity Ordered', ascending = False)
analisis_de_productos
```
